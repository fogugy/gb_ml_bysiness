```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np 
import pandas as pd 
import seaborn as sns

from sklearn.metrics import mean_squared_error
```

```python
df_duq = pd.read_csv('./data/DUQ_hourly.csv', index_col=[0], parse_dates=[0])
```

**Данные идут в неверном порядке. Отсортируем по индексу.**

```python
df_duq = df_duq.sort_index()
```

```python
df_duq.head()
```

```python
df_duq.describe()
```

```python
df_duq.info()
```

```python
df_duq.plot(style='.',figsize=(15,7),title='DUQ_MW потребление энергии')
plt.ylabel('МВт')
```

```python
def plot_for_year(year):
    mask = (df_duq.index >= pd.to_datetime(f'{year}')) & (df_duq.index < pd.to_datetime(f'{year+1}'))
    df_duq[mask].plot(style='.',figsize=(15,7),title=f'DUQ_MW потребление энергии {year}')
    plt.ylabel('МВт')
    plt.xlabel('')
```

```python
plot_for_year(2006)
```

```python
plot_for_year(2010)
```

- Основные статистики показали, что пропусков нет
